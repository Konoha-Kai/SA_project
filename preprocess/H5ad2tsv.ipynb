```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# 关闭警告
import warnings
warnings.filterwarnings("ignore")
from utils_type import load_h5ad
```

```python
gene_expression1,sc_meta1, cell_Specimen_ID1 = load_h5ad(input_bulk_path=r"/data/zhaoyh/output_fixed.h5ad",
                           input_sc_h5ad_path=r'/data/zhaoyh/output_fixed.h5ad',
                           input_st_data_path=r'/data/zhaoyh/output_fixed.h5adv',
                           input_st_meta_path=r'/data/zhaoyh/output_fixed.h5ad')
```

```python
gene_expression2,sc_meta2, cell_Specimen_ID2 = load_h5ad(input_bulk_path=r"/data/zhaoyh/test_output_fixed.h5ad",
                           input_sc_h5ad_path=r"/data/zhaoyh/test_output_fixed.h5ad",
                           input_st_data_path=r'/data/zhaoyh/test_output_fixed.h5ad',
                           input_st_meta_path=r'/data/zhaoyh/test_output_fixed.h5ad')
```

```python
gene_expression1 = gene_expression1.T
gene_expression2 = gene_expression2.T
gene_expression1
```

```python
# 自然连接两个dataframe
sc_meta1 = pd.merge(sc_meta1, cell_Specimen_ID1, on='Cell')
# 将Cell列变成索引,且保留原来的列
sc_meta1 = sc_meta1.set_index('Cell', inplace=None)
# 将原来的索引变成列
sc_meta1 = sc_meta1.reset_index()
sc_meta1
```

```python
sc_meta2 = pd.merge(sc_meta2, cell_Specimen_ID2, on='Cell')
sc_meta2 = sc_meta2.set_index('Cell', inplace=None)
sc_meta2 = sc_meta2.reset_index()
sc_meta2
```

```python
# 将两个dataframe合并
sc_meta = pd.concat([sc_meta1, sc_meta2])
sc_meta = sc_meta.reset_index(drop=True)
sc_meta
```

```python
# 将两个dataframe合并
gene_expression = pd.concat([gene_expression1, gene_expression2])
gene_expression
```

```python
# 设置gene_expression索引的名字
gene_expression.index.name = 'Gene_sample'
gene_expression
```

```python
# 重新设置sc_meta的列名
sc_meta.columns = ['Cell_ID', 'Cell_type', 'Sample_ID']
# 将sc_meta的Cell_ID列变成索引
sc_meta = sc_meta.set_index('Cell_ID', inplace=False)
sc_meta
```

```python
# 导出为tsv
gene_expression.to_csv(r'UMI_sc_gene\gene_expression.tsv', sep='\t')
sc_meta.to_csv(r'UMI_sc_gene\sc_meta.tsv', sep='\t')
```
